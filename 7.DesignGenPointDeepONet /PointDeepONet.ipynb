```python
import os

def patch_binary_vtk_header(src, dst=None, scan_bytes=5 * 1024 * 1024):
    """
    Patch non-standard legacy VTK dtype tokens (e.g., vtktypeint32) inside a BINARY .vtk
    by editing only the first `scan_bytes` bytes and keeping byte-length identical via padding.
    """
    if dst is None:
        dst = src.replace(".vtk", ".patched.vtk")

    mapping = {
        b"vtktypeint32":   b"int",          # 3
        b"vtktypeuint32":  b"unsigned_int", # 12
        b"vtktypeint64":   b"long",         # 4
        b"vtktypeuint64":  b"unsigned_long",
        b"vtktypefloat32": b"float",
        b"vtktypefloat64": b"double",
    }

    with open(src, "rb") as f:
        head = f.read(scan_bytes)
        tail = f.read()

    for k, v in mapping.items():
        if k in head:
            # pad to same length so we don't mess with any offsets/streaming
            v_padded = v + b" " * (len(k) - len(v))
            head = head.replace(k, v_padded)

    with open(dst, "wb") as f:
        f.write(head)
        f.write(tail)

    return dst

src = "/workspace/Point-DeepONet/data/VolumeMesh/6_247.vtk"
dst = patch_binary_vtk_header(src)
print("patched:", dst)

import pyvista as pv

m = pv.read(dst)
print(m)
print("n_points, n_cells:", m.n_points, m.n_cells)
print("point arrays:", list(m.point_data.keys())[:20])
print("cell arrays:", list(m.cell_data.keys())[:20])
```
